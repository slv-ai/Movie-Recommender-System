```python
#Dowload data 
def download_movielens_data(size='small'):
    """
    Download MovieLens dataset
    size: 'small' (100K ratings) or 'full' (25M ratings)
    
    For demo: use 'small' (fast, free)
    """
    import urllib.request
    import zipfile
    
    if size == 'small':
        url = 'http://files.grouplens.org/datasets/movielens/ml-latest-small.zip'
        filename = 'ml-latest-small.zip'
    else:
        url = 'http://files.grouplens.org/datasets/movielens/ml-latest.zip'
        filename = 'ml-latest.zip'
    
    print(f"Downloading {size} MovieLens dataset...")
    urllib.request.urlretrieve(url, filename)
    
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall('.')
    
    print("✓ Download complete")
    return filename.replace('.zip', '')
```

```python
data_dir = download_movielens_data('small')
    
    # Load ratings
print("\nLoading ratings...")
ratings = pd.read_csv(f'{data_dir}/ratings.csv')
movies = pd.read_csv(f'{data_dir}/movies.csv')
    
print(f"Total ratings: {len(ratings)}")
print(f"Total movies: {len(movies)}")
print(f"Total users: {ratings['userId'].nunique()}")
```
